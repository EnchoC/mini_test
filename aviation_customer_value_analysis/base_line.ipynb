### 2、数据清洗

```python
data['SUM_YR_1'].notnull()*data['SUM_YR_2'].notnull() # 找到两年总票价都不为零的记录
data = data[data['SUM_YR_1'].notnull()*data['SUM_YR_2'].notnull()]
print data_bak.shape
print data.shape
print 'delete data len[%d]'%(data_bak.shape[0]-data.shape[0])
```

### 3、属性规约

```python
# 与LRFMC模型相关的列
# FFP_DATE、LOAD_TIME、FLIGHT_COUNT、AVG_DISCOUNT、SEG_KM_SUM、LAST_TO_END
save_cols = ['FFP_DATE','LOAD_TIME','FLIGHT_COUNT',
             'avg_discount','SEG_KM_SUM','LAST_TO_END']
data_s = data[save_cols]
save_cols[3] = save_cols[3].upper()
data_s.columns = save_cols
```

### 4、数据变换

```python
merge_df.loc[['FFP_DATE','LOAD_TIME','FLIGHT_COUNT',
             'avg_discount','SEG_KM_SUM','LAST_TO_END']]
```

```python
# 转换数据格式
for col in data_s.select_dtypes(include='object').columns.tolist():
    data_s[col] = pd.to_datetime(data_s[col], format='%Y-%m-%d %H:%M:%S')
data_s.dtypes
```

```python
# L 会员人会时间距观测窗口结束的月数
# 观测窗口的结束时间 - 人会时间
# [单位:月]
data_s['L_score'] = (data_s['LOAD_TIME']- data_s['FFP_DATE']).dt.days/30
```

```python
# R 客户最近一次乘坐公司飞机距观测窗口结束的月数
# 最后一次乘机时间至观察窗口末端时长
# [单位:月]
data_s['R_score'] = data_s['LAST_TO_END']/30
```

```python
# F 客户在观测窗口内乘坐公司飞机的次数
# 观测窗口的飞行次数
# [单位:次]
data_s['F_score'] = data_s['FLIGHT_COUNT']
```

```python
# M 客户在观测时间内在公司累计的飞行里程
# 观测窗口的总飞行公里数
# [单位:公里]
data_s['M_score'] = data_s['SEG_KM_SUM']
```

```python
# C 客户在观测时间内乘坐舱位所对应的折扣系数的平均值
# 平均折扣率
# [单位:无]
data_s['C_score'] = data_s['AVG_DISCOUNT']
```

```python
data_tra = data_s.iloc[:,6:]
```

```python
data_tra.describe(percentiles=[])
```

```python
# 进行标准差标准化
data_nor = (data_tra - data_tra.mean(axis=0))/(data_tra.std(axis=0))
data_nor.head()
```

### 5、模型构建

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k = 5
kmodel = KMeans(n_clusters=k, n_jobs=4, random_state=20)
kmodel.fit(data_nor)
```

```python
cluster_count = pd.DataFrame(pd.DataFrame(kmodel.labels_)[0].value_counts()).sort_index()
cluster_center = pd.DataFrame(kmodel.cluster_centers_)
data_cluster = pd.concat([cluster_count,cluster_center], axis=1)
data_cluster.columns = ['count','ZL','ZR','ZF','ZM','ZC']
```

```python
data_cluster
```

```python
N = len(kmodel.cluster_centers_[0])  # 聚类中心的属性长度
angles=np.linspace(0, 2*np.pi, N, endpoint=False) # 设置雷达图的角度，用于平分切开一个圆面
angles=np.concatenate((angles, [angles[0]])) # 为了使雷达图一圈封闭起来
fig = plt.figure(figsize=(7,7)) # 设置画布大小
ax = fig.add_subplot(111, polar=True) # 这里一定要设置为极坐标格式
sam = ['r-', 'o-.', 'g--', 'b-.', 'p:'] # 样式
lab = [] # 图例标签名
for i in range(len(kmodel.cluster_centers_)):
    values = kmodel.cluster_centers_[i]
#     feature = ['ZF','ZC','ZR','ZL','ZM'] # 设置各指标名称
    feature = data_cluster.columns.tolist()[1:]
    # 为了使雷达图一圈封闭起来，需要下面的步骤
    values=np.concatenate((values,[values[0]]))
    ax.plot(angles, values, sam[i], linewidth=2) # 绘制折线图
    ax.fill(angles, values, alpha=0.25) # 填充颜色
    ax.set_thetagrids(angles * 180/np.pi, feature) # 添加每个特征的标签
    ax.set_ylim(-3, 3) # 设置雷达图的范围
    plt.title('Customer group feature map') # 添加标题
    ax.grid(True) # 添加网格线
    lab.append('Customer[%s]'%str(i+1))
plt.legend(lab, loc=4)
```
