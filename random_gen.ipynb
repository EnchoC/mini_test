```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import random
def rand5():
    return random.randint(0,6)
```

```python
def rand0_1():
    x = rand5()
    if x > float(5.0/2.0):
        return 1
    else:
        return 0
```

```python
def rand7():
    return sum([rand0_1() for x in xrange(0,8)])
```

```python
def binary_mapping():
    get_result = False
    while not get_result:
        x = rand5()
        if x in [0, 1, 2]:
            return 1
        elif x in [4, 5, 6]:
            return 0

def rand7_v2():
    return int(''.join([str(binary_mapping()) for x in xrange(0,3)]),2)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure()
plt.xlim(0,1000)
plt.ylim(0,8)
plt.plot([rand7_v2() for x in xrange(1,1000)])
```
